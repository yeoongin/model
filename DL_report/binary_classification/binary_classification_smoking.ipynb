```python
# 라벨링 파일 데이터 프레임 형태로 가져오기
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/DL_report_data/binary_calssification_smoking/smoking.csv')
```

```python
df
```

```python
df.shape
```

```python
X = df.iloc[:,0:26]
y = df.iloc[:,26]
```

```python
X.head
```

```python
# 문자형 데이터를 전부 숫자로 인코딩 해줌

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

categorical_features = ['gender', 'oral', 'tartar']

for col in categorical_features:
  data = X[col].values
  encoder.fit(data)       # 훈련 데이터를 기준으로

  X[col] = encoder.transform(data)
```

```python
X.head
```

```python
# 모든 피처들의 데이터 분포나 범위를 동일하게 조정해주기 위해 스케일링 적용
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
Scaled_X = pd.DataFrame(ss.fit_transform(X), columns=X.columns)
```

```python
# 학습셋과 테스트셋을 나눈다.
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(Scaled_X, y, random_state=2023, test_size=0.2)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
import tensorflow as tf
from keras.optimizers import Adam
from keras.callbacks import ModelCheckpoint, EarlyStopping

EPOCH = 500
BATCH_SIZE = 32

# 모델 네트워크 구성
model = Sequential()
model.add(Dense(128, input_dim=26, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# 우리가 풀어야할 문제는 이진분류이기 때문에 loss함수를 binary_crossentropy로 지정
model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])

model.summary()
```

```python
filename = 'checkpoint-epoch-{}-batch-{}-trial-001.h5'.format(EPOCH, BATCH_SIZE)

# 가장 best인 모델을 저장하기 위함
checkpoint = ModelCheckpoint(filename,             # file명을 지정
                             monitor='loss',       # loss 값이 개선되었을때 호출
                             verbose=1,            # 로그를 출력
                             save_best_only=True,  # 가장 best 값만 저장
                             mode='auto'           # auto는 알아서 best를 찾음
                            )

# 과적합 방지를 위한 하나의 방법
# patience 횟수만큼 monitor로 설정한 모델의 성능이 개선되지 않으면 자동으로 학습을 종료함

earlystopping = EarlyStopping(monitor='loss',  # 모니터 기준 설정 (loss) -> 정확도로도 설정할 수 있음
                              patience=10,         # 10회 Epoch동안 개선되지 않는다면 종료
                             )

# 모델을 실행한다.
model.fit(X_train, y_train, epochs=EPOCH, batch_size=BATCH_SIZE, callbacks=[checkpoint, earlystopping])
```

```python
# 테스트 결과를 출력한다.

from keras.models import load_model
load_model = load_model('checkpoint-epoch-500-batch-32-trial-001.h5')
score = load_model.evaluate(X_test, y_test)

print('Test accuracy: ', score[1])
```
