```python
X = df.iloc[:,0:11]
y = df.iloc[:,11]
```

```python
X.head
```

```python
# 문자형 데이터를 전부 숫자로 인코딩 해줌

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

categorical_features = ['gender']

for col in categorical_features:
  data = X[col].values
  encoder.fit(data)       # 훈련 데이터를 기준으로

  X[col] = encoder.transform(data)
```

```python
X.head
```

```python
#원핫 인코딩 처리
y = pd.get_dummies(df['class'])
```

```python
y
```

```python
# 모든 피처들의 데이터 분포나 범위를 동일하게 조정해주기 위해 스케일링 적용
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
Scaled_X = pd.DataFrame(ss.fit_transform(X), columns=X.columns)
```

```python
# 학습셋과 테스트셋을 나눈다.
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(Scaled_X, y, random_state=2023, test_size=0.2)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
import tensorflow as tf
from keras.optimizers import Adam
from keras.callbacks import ModelCheckpoint, EarlyStopping

EPOCH = 500
BATCH_SIZE = 8

# 모델 구성
model = Sequential()
model.add(Dense(64, input_shape=(11, ), activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(4, activation='softmax'))
'''
다중분류이기 때문에 마지막 Dense Layer는 활성화 함수를 softmax를 사용
노드 수는 우리가 맞추고자하는 정답 클래스 개수랑 같아야함 (6개)
'''

# 현재 레이블이 정수(int)형태를 갖는 다중 분류이기 때문에 loss함수를 sparse_categorical_crossentropy로 지정
# 레이블 인코딩이아니라 원핫 인코딩일 경우 categorical_crossentropy로 지정해야함
model.compile(loss='categorical_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])

model.summary()
```

```python
# 저장할 파일 이름 지정
filename = f'checkpoint-epoch-{EPOCH}-batch-{BATCH_SIZE}-trial-001.h5'

# 가장 best인 모델을 저장하기 위함
checkpoint = ModelCheckpoint(filename,             # filename명 지정
                             monitor='loss',       # loss값이 개선되었을때 저장
                             verbose=1,            # 로그를 출력
                             save_best_only=True,  # 가장 best 값만 저장
                             mode='auto'           # auto는 알아서 best를 찾음
                            )


# 과적합 방지를 위한 하나의 방법
# patience 횟수만큼 monitor로 설정한 모델의 성능이 개선되지 않으면 자동으로 학습을 종료함

earlystopping = EarlyStopping(monitor='loss',       # 모니터 기준 설정 (loss) -> 정확도로 설정할 수 있음
                              patience=10,          # 10회 Epoch동안 개선되지 않는다면 종료
                             )

# 모델 학습
model.fit(X_train, y_train, epochs=EPOCH, batch_size=BATCH_SIZE, callbacks=[checkpoint, earlystopping])
```

```python
from keras.models import load_model
load_model = load_model('checkpoint-epoch-500-batch-8-trial-001.h5')
score = load_model.evaluate(X_test, y_test)

print('Test accuracy: ', score[1])
```
